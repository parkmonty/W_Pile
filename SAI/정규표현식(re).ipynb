# **정규표현식(re)**

## 1. 정규 표현식 (RE : Regular Expression)

* 특정한 규칙을 가진 문자열 패턴을 사용하는 방식
* 데이터 전처리에 있어 텍스트에서 특정 문자열을 검색, 치환, 제거할 때 사용한다.

## 2. 정규 표현식 사용

```python
import re
```

re.compile("")은 정규식 객체를 리턴해준다.

      p = re.compile("패턴")

image.png

* [  ] : 대괄호

  * 대괄호 사이에 원하는 문자를 여러 개 넣으면, 문자열이 넣은 문자 중 하나와 일치하면 매칭이 된다. (OR 개념)
  * 즉, 대괄호 안에 있는 문자 중 하나만 일치해도 된다.
  * 대괄호 안에서 . * + 는 하나의 문자가 된다.

* ^ : 캐릿

  * 대괄호 앞에 ^이 들어가면 대괄호의 내용과 반대되는 것을 반환
  * ^[0-9] -> 숫자가 아닌것을 반환

* . : 마침표

  * 모든 문자와 일치하는 (개형을 제외) 정규표현식
  * 단, [  ] 안에 들어가 있으면 하나의 . 로 취급

## 3. 정규식 그룹

* () 괄호로 그룹을 묶을 수 있다.
* (\d)-(\d-\d) 으로 묶었다면 group(1)은 \d, group(2)는 \d-\d가 된다.

## 4. 그외 정규식

* re.match('패턴', '검사하려는 문자열') : 문자열의 처음부터 패턴 일치 여부 확인

* re.fullmatch() : 문자열 전체가 해당 패턴인지를 검사

* re.search : 처음부터가 아니라 중간에 있어도 상관 없음.

* re.findall : 문자열 중 패턴과 일치되는 모든 부분 찾기

* re.sub() : 문자열에서 정규 표현식과 일치하는 부분에 대해서 다른 문자열로 대체

* re.finditer() : 문자열에서 정규 표현식과 매치되는 모든 경우의 문자열에 대한 이터레이터 객체를 리턴

* re.split() : 정규 표현식을 기준으로 문자열을 분리하여 리스트 리턴
