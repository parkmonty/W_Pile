# 데이터 분석에 필요한 기초

```python
# 리스트
a = ['사과', '딸기', '수박', '참외', '파인애플', '멜론']
# 2차원 리스트
b = [['사과', '딸기'], ['수박', '참외'], ['파인애플', '멜론']]

# 리스트를 접근하는 방법
for i in a:
  print(i)
print('----')
for j in range(len(a)):
  print(a[j])

# 2차원 리스트 접근 방법
print('\n<2차원 리스트>\n')

for i in range(len(b)):
  for j in range(len(b[i])):
    print(b[i][j])
```

*   2차원
*   가로와 세로 라는 공간
*   행렬의 개념
*   인덱스의 개념과 동일하게 접근이 되는 것
*   (0,0)(0,1)
*   (1,0)(1,1)
*   (2,0)(2,1)

*   데이터프레임의 개념으로 접근
*   2차원의 개념으로 행과 열
*   행은 인덱스, 열은 컬럼이 된다.


*   데이터프레임을 불러오기 위해서는
