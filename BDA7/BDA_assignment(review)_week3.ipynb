### 1-1. 양적변수

* 변수를 숫자로 나타 낼 수 있는 변수
* 양을 표현하는 것, 학교 시험 점수, 키, 몸무게 등
* 연산이 가능

#### 1-1-1. 구간(등간) 척도 (interval scale)

* 연속적인 수로 수량화할 수 있음
* 원점이 존재하지 않음
* 어느 정도 크고 작다를 표현할 수 있음
* 숫자 간의 간격이 동일
* '0'이라는 값이 값이 없다고 할 수 없다.

ex) 절대적인 원점이 없기 때문에 온도 0도는 온도가 없다고 할 수 없다.

#### 1-1-2. 비율 척도 (ratio scale)

* 연속적인 수로 수량화할 수 있음
* 절대적인 원점이 존재
* '0' 값이 의미 있는 값을 나타내며 이 값은 해당 수량의 완전한 부재를 나타냄

ex) 절대 영점이 존재하기 때문에 몸무게 0kg은 몸무게가 없다고 할 수 있다

### 1-2. 질적변수

* 변수의 값이 자료를 특정 카테고리에 포함 시키도록 하는 변수
* 숫자로 표현할 수 없는 정보로 그룹으로 묶기 때문에 범주형 자료라고 함
* 연산이 불가능

#### 1-2-1. 명목 척도 (nominal scale)

* 이름만 나타내는 척도
* 숫자로 표현이 가능하나 숫자는 분류의 기능만 할 뿐 양적인 의미는 없고 특정카테고리만 표현

ex) 취미 (운동 0, 독서 1, 미술 2)

#### 1-2-2. 서열 척도 (ordinal scale)

* 순위(서열)이 있는 척도
* 수량화할 수 있는 자료
* 숫자는 서열을 나타내지만 '어느 정도' 크고 작다는 것은 표현 불가능

ex) 달리기 순위 (1위, 2위, 3위), 만족도 (높은 2, 보통 1, 낮음 0)

### 1-3. 이산변수

* 이산 할 수 있는 변수를 의미
* 하나하나 셀 수 있는 변수를 의미
* 정수의 범위

ex) 아파트 층 수, 한 회사 직원의 수 등

### 1-4. 연속변수

* 이산변수와 다르게 변수의 각 값 사이네 무수히 많은 또 다른 값들이 존재하는 경우
* 실수의 범위

ex) 사람의 키, 몸무게, 한 가구의 소득 등

### 1-5. 이산형, 연속형을 가지고 바라보는 ML, 통계적인 분석

* 머신러닝 - 회귀, 분류
* 통계분석 - 가설검정, 정규성, 순서관련한 다양한 통계적 분석, 감정들
* 변수에 따라서 -> 척도에 따라서 -> 이산형, 연속형에 따라서 접근을 다르게 해야 함

* 평균의 종류도 다양하다.

#### 2-1-1. 산술평균

* 일상 생활에서 가장 많이 사용하는 평균
* 각 요소의 총 합을 갯수로 나눈 것
* **산술평균 = (a + b) / 2**

#### 2-1-2. 기하평균

* 각 요소를 곱한후 그 값에 루트를 씌운 값
* 일반적으로 성장률의 평균 등 상승률과 하락률을 활용한 계산에 유용하게 쓰임
* **기하평균 = sqrt(a * b)**

* 예를 들어 1,000원의 주식이 10% 상승 후 10% 하락 했을 경우 산술적인 개념만 본다면 상승 +10% 하락 -10%로 ±10%이기 때문에 직관적으로 잘못 생각하여 가격 변동이 없다고 착각할 수 있음
* 하지만, 1,000원이 10% 상승하면 → 1,100원이 되고 1,100원이 10% 하락하면 → 990원이 됩니다
* 위 내용처럼 1,000원 짜리 주식이 990원이 되는 것을 상승률 10% 와 하락률 10%에 대한 기하 평균을 활용하여 계산할 수 있음
* 수식으로 표현해보면 (1.1 x 0.9)^0.5 = 0.99 → 1,000원 x 0.99원 = 990 라는 결과를 얻을 수 있다.

#### 2-1-3. 조화평균

* 각 요소의 역수를 산술평균한 우 그 값을 다시 역수로 변환한 것
* **조화평균 = 2 * a * b / (a + b)**

* 예를 들어 두 지점 A, B에 대하여 갈때는 시속 10Km/h, 올때는 시속 20Km/h로 왕복했다면, 평균속력은 조화평균인 13.33Km/h가 된다.
* ((1/10 + 1/20) / 2) ^ (-1) = 40/3 = 13.33

#### 2-1-4. 산술평균 >= 기하평균 >= 조화평균

* 등호는 a = b 일 때 성립한다.

* Z스코어는 값 그룹의 평균에 대한 값의 관계를 설명하는 수치 측정
* Z스코어는 평균으로부터의 표준 편차로 측정
* Z-score가 0이면 데이터 포인트의 점수가 평균 점수와 동일함을 나타낸다
* 1의 Z스코어는 평균에서 1 표준 편차인 값을 나타낸다
