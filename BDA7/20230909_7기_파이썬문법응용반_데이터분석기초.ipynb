```python
a= ['파문응','파문기','데분기','데분중']
```

```python
#2차원 리스트?
b =[['파문응','파문기'],['데분기','데분중']]
```

```python
#리스트를 접근하는 방법?
#리스트 요소를 가지고 오기 위해서는?
for i in a:
    print(i)
print('----')
for i in range(len(a)):
    print(a[i])
```

- 2차원
- 가로와 세로 라는 공간
- x축과 y축의 공간
- 행렬의 개념
- x,y 좌표 생각을 하면
- 인덱스의 개념과 동일하게 접근이 되는 것
- 0,0 , 0,1
- 1,0 , 1,1

- 데이터프레임의 개념으로 접근하면
- 2차원의 개념으로 행과 열

- 컬럼이 존재하는 건 결국 데이터의 성질이 형성이 된다.
- 우리가 저번에 배운 데이터에 대한 개념을 이해하고 생각하면 해당 컬럼이 데이터의 특징을 나타내는 것

```python
# 패키지를 이용해서 데이터를 불러와 보자!
df = sns.load_dataset('titanic')
```

```python
df[['survived','pclass','sex','fare','class']]
# 타이타닉데이터는 생존 유무를 예측하거나
# 생존이라는 컬럼 survived 종속변수의 개념으로 y값으로 생각하는 것 라벨

# 변수가 많다는 게 모든 걸 다 설명하는 것일까?
# 변수가 적어도 충분히 다 설명할 수 있는 게 아닌가?
# 이런 것들을 가지고 변수 선택에 대한 고민과 기준을 배울 예정

# 차원이 많아지면서 설명하기 어려워지고 , 희소한 행렬과 변수관의 상관성도 생기고 등등..
# 꼭 적은 변수더라도 이 변수가 설명할 수 있는 생존에 대한 컬럼을 잘 분류할 수 있는 컬럼이라면 충분히 좋은 성능이 나올 수 있다.

# 그 기준은 무엇인가?
# 모델링 ML으로 생존 예측을 한다면-> 결국 정답은 생존? 생존 y값에 대해서 얼마나 다른 변수들이 잘 구분을 할 수 있는가? 이것이 중요한 선택의 요인 중 하나
# 데이터 분석가의 도메인이 가장 중요합니다.
# 분석가의 도메인을 제외한 기술적인, 통계적인 변수 선택은 y값에 구분을 잘 할 수 있는가?

# 파문응 160명 있음
# 파문응 7기의 우수학회원을 가장 잘 분류할 수 있는 컬럼은 무엇일까?
# 파문응 우수학회원이 15명만 남음
# 변수를 선택한다면?
# 성별 컬럼, 수업 참여 컬럼
# 남/여
# 15명 중 7남/8여
# 수업 참여
# 모든 수업 올 참여

# 전체 데이터를 기준으로 보면
# 전체
# 파문응의 전체 인원 160
# 남 여 80//80 -> 전체 값을 분류하기에는 좀 어렵다.
# 우수학회원은 모든 수업참여 all 나머지 학회원은 0 가까우면
# 우수학회원 특징은 수업참여를 대부분 다 모두 all
# 우수학회원을 분류할 수 있는 중요한 피처라는 것
```
