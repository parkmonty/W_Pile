# 실습 문제

```python
import pandas as pd
```

```python
# 각 줄마다 이름, 성, 생일년도 가 적힌 csv 파일이 있다. 이 파일을 읽어 chart라는 데이터프레임을 만들어라
chart = pd.read_csv("인적사항.csv")
print(chart)
```

```python
# "생일년도" 열에 비어있는 값이 있다. chart와 동일하지만 비어있는 값이 1980.0로 채워진 fullChart 데이터프레임을 만들고 출력하라
chart['생일년도'] = chart['생일년도'].fillna(1980.0)
fullchart = chart
fullchart
```

```python
# fullChart의 "생일년도" 열에 저장된 고유값들을 출력하라
fullchart['생일년도'].unique()
```

```python
# apply() 함수를 이용하여 fullChart의 "생일년도" 자료형을 정수로 바꾼 뒤, 출력하라
fullchart['생일년도'] = fullchart['생일년도'].apply(int)
fullchart
```

```python
# applymap()을 이용하여 fullChart 각 열에 저장된 정보의 자료형이 무엇인지 알려주는 데이터프레임 chartDataType을 만들고 다음처럼 출력하라
#               이름             성별             생일년도
# 0  <class 'str'>  <class 'str'>  <class 'int'>
chartDataType = fullchart[:1].applymap(lambda x : type(x))
chartDataType
```

```python
# map() 함수를 이용하여 fullChart의 "성별" 열에 man을 "남", woman을 "여"로 바꾼 뒤 출력하라
import numpy as np

fullchart['성별'] = np.where(fullchart['성별'] == 'man', '남', '여')
fullchart

# fullchart['성별'] = fullchart['성별'].map({'man' : '남', 'woman' : '여'})
```

```python
# fullChart에서 남, 여가 각각 몇 명 있는지 출력하여라
fullchart['성별'].value_counts()
```

```python
import pandas as pd

student_list = [{'이름' : '김현우', '단과대' : '소융대', '학번' : 23},
                {'이름' : '박민서', '단과대' : '소융대', '학번' : 21},
                {'이름' : '박민지', '단과대' : '소융대', '학번' : 22},
                {'이름' : '이준희', '단과대' : '소융대', '학번' : 20},
                {'이름' : '전준호', '단과대' : '생과대', '학번' : 21},
                {'이름' : '최윤제', '단과대' : '소융대', '학번' : None},
               {'이름' : '박민서', '단과대' : '자연대', '학번' : 22}]


#각 문제별로 지시하는 사항은 항상 df에 반영(저장)되어야 합니다.

#1번. student_list와 columns를 사용해 DataFrame을 만드세요.
# columns = ['이름', '단과대', '학번']
df = pd.DataFrame.from_dict(student_list)
df
```

```python
#2번. 한 명의 이름이 두 번 기입되었습니다. 중복된 이름 중 처음으로 나오는 값을 지우세요.
df.drop_duplicates(['이름'], inplace = True)
df
```

```python
#3번. 소융대 학생 중 한 명의 학번이 입력되지 않았다. 이 Nan값을 찾아 소융대에서 가장 높은 (숫자가 작은) 학번으로 바꾸세요.
# 단, filla와 groupby를 사용해야 함

num_list = df.groupby("학번")
name_list = []

for name, group in num_list:
  name_list.append(int(name))

min_num = min(name_list)

df.fillna(min_num, inplace = True)
df
```

```python
#4번. 모두 20살에 입학했다고 가정했을 때, 학번으로 나이를 구하는 함수를 작성하세요.

#5번. 4번에서 만든 함수를 이용해 각 학생의 나이를 구하고 '나이' 열을 추가해 저장하세요.
# 단, apply를 사용해야 함

def age(df):
  age_list = 23 - df + 20
  return age_list


df['나이'] = df['학번'].apply(age).map(int)
df
```
