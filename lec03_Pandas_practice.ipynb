## 2. 데이터 합치기 (concat, merge)

### 2.1. 행 기준으로(위아래로) 합치기

```python
df1, df2 = df.loc[:400], df.loc[400:]
```

```python
df1
```

```python
df2
```

- concat 함수를 테스트하기 위해 원본 데이터프레임을 임의로 나눔

```python
df_concat = pd.concat([df1, df2])
df_concat
```

- concat 함수는 단순히 두 데이터프레임을 위아래로 연결

### 2.2. 열 기준으로(옆으로) 합치기

```python
df1, df2 = df.iloc[:, :7], df.iloc[:, 7:]
df1.insert(0, 'passenger_id', df.index)
df2.insert(0, 'passenger_id', df.index) # 테스트를 위한 코드. 이해할 필요 X, 실행만 O.
```

```python
df1
```

```python
df2
```

- 대다수의 빅데이터는 다양한 정보가 서로 다른 데이터 테이블에 담겨 있음
- 메타 데이터를 병합하는 과정이 필요
- 원본 데이터셋이 두 가지로 나뉜 데이터 테이블이었다고 가정하기 위해 임의로 나눔

```python
df_concat = pd.concat([df1, df2], axis=1)
df_concat
```

- concat 함수의 파라미터에 axs=1 로 설정하면 단순히 두 데이터프레임을 나란이 옆에 이어줌
- passenger_id가 두 번 등장

```python
df_merge = pd.merge(df1, df2, on='passenger_id', how='left')
df_merge
```

- merge 함수를 사용하면 SQL 문의 join과 동일한 동작 가능
- key 값을 passenger_id로 하여 join 하였으므로, 두 데이터프레임의 order가 달라도 passenger_id의 칼럼이 같은 값을 가진다면 key 값을 기준으로 병합 가능

## 4. 결측값 처리하기

```python
df.isnull().sum() # 칼럼별 결측값 개수 확인
```

```python
df.dropna() # 하나라도 결측값 존재하면 그 행을 삭제
```

```python
df = df.dropna(subset='embark_town') # subset 파라미터로 어떤 칼럼을 기준으로 결측값을 삭제할 것인지 설정 가능
df
```

```python
df['age'] = df['age'].fillna(df['age'].mean()) # 결측값이 있을 때 새로운 값으로 채워서 처리
df
```

- 결측값을 새로운 값으로 채울 때는 사용자가 임의의 값을 설정하거나 평균, 중앙값 등의 대표값을 사용

## 7. apply 함수

```python
def change_class_name(class_name):
  if class_name == 'First':
    return '1st'
  elif class_name == 'Second':
    return '2nd'
  else:
    return '3rd'
```

```python
df['class'] = df['class'].apply(change_class_name)
df
```

- 행 단위로 반복 작업을 수행할 때, 함수를 만든 뒤 apply를 사용해 일괄 적용시킬 수 있음

## 8. 문자열 처리 함수 str
- type이 문자열인 칼럼에 대해서만 사용 가능

```python
df['embark_town'].str.contains('South') # South가 포함된 문자열인지
```

```python
df[df['embark_town'].str.contains('South')] # 위의 논리식(또는 조건)을 적용해 원하는 문자열이 포함된 행만 출력할 수 있음
```

```python
df['embark_town'] = df['embark_town'].str.upper() # 대문자로 바꾸기
df
```

```python
df['embark_town'] = df['embark_town'].str.lower() # 소문자로 바꾸기
df
```

- 이외에도 str 함수는 공백을 기준으로 문자열을 나누는 split, 합치는 strip, 특정 문자열을 대체하는 replace 등 다양한 함수 존재
- 텍스트 데이터를 pandas로 다룰 때 편리하게 쓰이는 함수
